```python
def result(Team, Opposition, Home_team, City, lighting):
    
    dataset = pd.read_csv('Match_predictions.csv')
    vectorized = pd.get_dummies(dataset.iloc[:,1:6])
    
    flag = [False, False, False, False, False]
    
    record = np.zeros((1,93),dtype='float64')
    t = "Team_" + Team
    o = "Opposition_" + Opposition
    h = "Home team_" + Home_team
    c = "City_" + City
    l = lighting
        
    record[0,92] = l
    
    for i,vec in enumerate(vectorized.columns):
        if vec == t:
            record[0,i] = 1
            flag[0] = True
        if vec == o:
            record[0,i] = 1
            flag[1] = True
        if vec == h:
            record[0,i] = 1
            flag[2] = True
        if vec == c:
            record[0,i] = 1
            flag[3] = True
            
    if(flag[0]==True and flag[1]==True and flag[2]==True and flag[3]==True):
        filename = 'model_winnerSVC.sav'
        model = pickle.load(open(filename, 'rb'))
        ans = model.predict(record)
        
        if ans == 1:
            print("INDIA will win")
        elif ans == 2:
            print("AUS will win")
            
    else:
        print("Incorrect details!")
```
